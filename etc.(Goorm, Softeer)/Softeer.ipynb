### 1) [플레이페어 암호](https://softeer.ai/practice/info.do?eventIdx=1&psProblemId=804)
- [공식 해설](https://softeer.ai/community/view.do?idx=693&cd=edu&pageNo=1)

```python
import sys
input = sys.stdin.readline

# 입력
message = list(input())
message = message[:-1]
message_key = list(input())

# 채워 넣기
map_size = 5
key_map = [[] for _ in range(map_size)]
# print(key_map)

ref = [
    'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N',
    'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z'
]

i = 0
key_map_1D = []
for m in message_key:
    if m in ref:
        key_map[i // map_size].append(m)
        ref.remove(m)
        i = i + 1
        key_map_1D.extend(m)

for mm in ref:
    key_map[i // map_size].append(mm)
    i = i + 1
    key_map_1D.extend(mm)
## 채워 넣는 것까지 완료

# 암호화
# 메세지 두 글자씩 자르고, 중복 글자 처리
temp_change = []
visited = [False] * len(message)
message_length = len(message)
for idx in range(message_length):
    if visited[idx]:
        continue

    if idx != message_length-1:
        if message[idx] != message[idx+1]:
            temp_change.append([message[idx], message[idx+1]])
            visited[idx] = True
            visited[idx+1] = True
        elif message[idx] == 'X' and message[idx+1] == 'X':
            temp_change.append([message[idx], 'Q'])
            visited[idx] = True
        else:
            temp_change.append([message[idx], 'X'])
            visited[idx] = True
    else:
        temp_change.append([message[idx], 'X'])
        visited[idx] = True
# 암호화 완료

# 마지막 두 글자 암호화
secrete_final = []
for temp_set in temp_change:
    char1 = temp_set[0]
    char2 = temp_set[1]

    char1_idx = key_map_1D.index(char1)
    char2_idx = key_map_1D.index(char2)

    char1_row = char1_idx // map_size
    char2_row = char2_idx // map_size

    char1_col = char1_idx % map_size
    char2_col = char2_idx % map_size

    if char1_row == char2_row:
        change_char1 = key_map[char1_row][(char1_col+1) % map_size]
        change_char2 = key_map[char2_row][(char2_col+1) % map_size]
        secrete_final.extend([change_char1, change_char2])
    elif char1_col == char2_col:
        change_char1 = key_map[(char1_row+1) % map_size][char1_col]
        change_char2 = key_map[(char2_row+1) % map_size][char2_col]
        secrete_final.extend([change_char1, change_char2])
    else:
        change_char1 = key_map[char1_row][char2_col]
        change_char2 = key_map[char2_row][char1_col]
        secrete_final.extend([change_char1, change_char2])
# 암호화 끝

for each in secrete_final:
    print(each, end='')
```
