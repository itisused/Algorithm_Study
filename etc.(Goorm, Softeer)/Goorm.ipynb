```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
# user_input = input()
# print ("Hello Goorm! Your input is " + user_input)
import sys
input = sys.stdin.readline
import heapq
import pdb

INF = int(1e9)

# 입력
temp = []
max_num = 0
while True:
	a, b = map(int, input().split())
	if a == 0 and b == 0:
		break
	# 노드 개수가 몇개인지 모르잖아 ㅆㄷ
	if a > b & a > max_num:
		max_num = a
	elif a < b & b > max_num:
		max_num = b
	else:
		max_num = max_num
	temp.append([a, b])

graph = [[] for _ in range(max_num+1)]
# visiteds = [[False] * (max_num+1) for _ in range(max_num+1)]
# distances = [[INF] * (max_num+1) for _ in range(max_num+1)]

for x, y in temp:
	graph[x].append((y, 1))
# print(graph, visited, distance)

'''
def get_smallest_node():
	min_value = INF
	index = 0 # 가장 최단 거리가 짧은 노드(인덱스)
	for i in range(1, n + 1):
		if distance[i] < min_value and not visited[i]:
			min_value = distance[i]
			index = i
	return index

def dijkstra(start):
	# 시작 노드에 대해서 초기화
	distance[start] = 0
	visited[start] = True
	for j in graph[start]:
		distance[j[0]] = j[1]
	# 시작 노드를 제외한 전체 n - 1개의 노드에 대해 반복
	for i in range(n - 1):
		# 현재 최단 거리가 가장 짧은 노드를 꺼내서, 방문 처리
		now = get_smallest_node()
		visited[now] = True
		# 현재 노드와 연결된 다른 노드를 확인
		for j in graph[now]:
			cost = distance[now] + j[1]
			# 현재 노드를 거쳐서 다른 노드로 이동하는 거리가 더 짧은 경우
			if cost < distance[j[0]]:
				distance[j[0]] = cost
'''

# def dijkstra(start, distance, visited):
def dijkstra(start, distance):
	q = []
	# 시작 노드로 가기 위한 최단 경로는 0으로 설정하여, 큐에 삽입
	heapq.heappush(q, (0, start))
	distance[start] = 0
	while q: # 큐가 비어있지 않다면
		# 가장 최단 거리가 짧은 노드에 대한 정보 꺼내기
		dist, now = heapq.heappop(q)
		# 현재 노드가 이미 처리된 적이 있는 노드라면 무시
		if distance[now] < dist:
			continue
		# 현재 노드와 연결된 다른 인접한 노드들을 확인
		for i in graph[now]:
			cost = dist + i[1]
			# 현재 노드를 거쳐서, 다른 노드로 이동하는 거리가 더 짧은 경우
			if cost < distance[i[0]]:
				distance[i[0]] = cost
				heapq.heappush(q, (cost, i[0]))
				# print("push: ", q, cost, i[0])
	# return distance, visited
	return distance

# for start, distance, visited in zip(range(1, max_num+1), distances, visiteds):
# 	dijkstra(start, distance, visited)
distances = []
# visiteds = []
result_sun = []
cnt = 0
for start in range(1, max_num+1):
	# pdb.set_trace()
	# 연결 노드 없는 그래프는 지나가기
	if not graph[start]:
		# result_sun.append(0)
		continue
	# else:
	cnt = cnt + 1

	# # 방문한 적이 있는지 체크하는 목적의 리스트를 만들기
	# visited = [False] * (n + 1)
	# 최단 거리 테이블을 모두 무한으로 초기화
	distance = [INF] * (max_num + 1)

	# distance, visited = dijkstra(start, distance, visited)
	distance = dijkstra(start, distance)
	# distance.replace(1e9, 0)
	distances.append(distance)
	# visiteds.append(visited)
	# result_sun.append(sum(distance[1:]))
	temp = 0
	for d in distance[1:]:
		if d == INF:
			continue
		temp = temp + d
	result_sun.append(temp)

total_mom = cnt * (cnt-1)
total_sun = sum(result_sun)
# print(result_sun)

answer = total_sun / total_mom

print(round(answer, 2))
```
