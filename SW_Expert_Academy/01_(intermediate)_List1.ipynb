### 8차시 - 숫자 카드
- numpy 못 씀
- labda 함수를 이용한 정렬! [참고](https://kingofbackend.tistory.com/98)

```python
for test_case in range(1, T+1):
    N = int(input()) # # of card
    non_sep_num = input()

    # 분리
    sep_num = {}
    for num in non_sep_num:
        num = int(num)
        if num in sep_num.keys():
            sep_num[num] = sep_num[num] + 1
        else:
            sep_num[num] = 1
    # print('중간확인1: ', sep_num)

    sep_list = list(sep_num.items())
    # print('중간확인2: ', sep_list)

    sep_list_sort = sorted(sep_list, key = lambda x: (x[1], x[0]))
    # print('중간확인3: ', sep_list_sort)
    
    # 결과는 sep_list_sort와 동일한 것으로 판단되나 SWEA에서 실행 시 Runtime Error 발생
    # sep_list.sort(key = lambda x: (x[1], x[0]))

    print('#{0} {1} {2}'.format(test_case, sep_list_sort[-1][0], sep_list_sort[-1][1]))
```

### 9차시 - 구간합

```python
for test_case in range(1, T+1):
    N, M = map(int, input().split())
    v = list(map(int, input().split()))

    # Greedy
    result = []
    for i in range(N-M+1): # 얘랑
        result.append(sum(v[i:i+M])) # 얘 좀 헷갈리지 말 것!
    
    answer = max(result) - min(result)
    print('#{0} {1}'.format(test_case, answer))
```
