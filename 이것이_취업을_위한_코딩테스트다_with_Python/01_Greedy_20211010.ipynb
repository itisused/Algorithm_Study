## Example 02: 큰 수의 법칙
- 책 92

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|:----:|
|21.10.09|30분 초과|X|X||

```python
# 1. 입력부
n, m, k = map(int, input().split(' '))
# temp = map(int, input().split(' ')) # n의 수 만큼 입력
temp = list(map(int, input().split(' ')))

# # Test 입력
# n, m, k = map(int, [5, 8, 3])
# temp = [2, 4, 5, 4, 6]

# 2. 동작부
# 흰트: 큰 수 2개만 추출
max1 = max(temp)
temp.remove(max(temp))
max2 = max(temp)
# temp.remove(max(temp))

# 덧셈
a = m // k
b = m % k

# max1을 전체 m번 중 k번씩 m//k번 반복하고 그 중간 사이사이 m%k번마다 max2를 더함
result = max1 * a * k + max2 * b

# 3. 출력부
print(result)

```

## Example 03: 숫자 카드 게임

|21.10.10|20분 소요|X|X||

```python
# 1. 입력부
# 행렬 크기 입력
m, n = map(int, input().split(' '))
# 행렬 입력 / 옳게 입력한다는 가정하에
mtx = []
for i in range(m):
    mtx.append(list(map(int, input().split(' '))))

# # Test 입력
# m, n = map(int, [3, 3])
# mtx = [[3, 1, 2], [4, 1, 4], [2, 2, 2]]

# 2. 동작부
min_value = []
for mtx_rows in mtx:
    min_value.append(min(mtx_rows))

print(max(min_value))
```

## Example 04: 1이 될 때까지
- 책 99 (2018 E 기업 알고리즘 대회)

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|:----:|
|21.10.10|5분 소요|X|X||

```python
# 1. 입력부
n, k = map(int, input().split(' '))

# 2. 동작부
cnt = 0
while n != 1:
    if n % k == 0:
        n = n // k
    else:
        n = n - 1
    cnt = cnt + 1
# 3. 출력부
print(cnt)
```

## etc.: 펭귄추락대책위원회
- 백준 [18228](https://www.acmicpc.net/problem/18228) (INU 송년 코드페스티벌 2019)

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|----|
|21.10.10|?(빠르게 끝남)|X|X|- Problem 06 대체 <br> - 거창해보이는 문제에 비해 내용은 쉬움(허무) <br> - "양옆으로 인접해 있는 얼음들을 하나의 그룹으로 봤을 때, <br>그룹의 끝이 얼음1 또는 N을 포함하지 않는다면 <br>이 얼음 그룹은 아래로 추락하게 된다."가 뭔소리야?|

```python
# 1. 입력부
n = int(input())
a_i = list(map(int, input().split(' ')))

# # Test 입력
# n = 5
# a_i = [7, -1, 6, 2, 5]

'''
[뭔 소리야?]
양옆으로 인접해 있는 얼음들을 하나의 그룹으로 봤을 때,
그룹의 끝이 얼음1 또는 N을 포함하지 않는다면
이 얼음 그룹은 아래로 추락하게 된다.
=> 양 끝을 동시에 선택하면 안된다? 인건가??
'''
# 2. 동작부
penguin = a_i.index(-1)

left_power = min(a_i[:penguin])
right_power = min(a_i[penguin+1:])

# 3. 출력부
print(left_power + right_power)
```
