## Problem 34: 병사 배치하기
- 백준 [18353](https://www.acmicpc.net/problem/18353)

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|----|
|21.12.07|40분 초과|o|o|- ...| 

```python
# 1. 입력
N = int(input())
solider = list(map(int, input().split()))

# 2. 동작
# 간격을 계산해서 (앞에 것 - 뒤에 것)
# 간격의 부호가 바뀌는 값을 제외한다?

gap = []
count = 0
for i in range(N-1):
    temp = solider[i] - solider[i+1]
    if temp > 0:
        gap.append(1)
    else:
        gap.append(0)
        count = count + 1

# 설마 이게 끝?
print(count)
```
