````python
# 방문하지 않은 노드 중에서, 가장 최단 거리가 짧은 노드의 번호를 반환
def get_smallest_node():
    # pdb.set_trace()
    min_value = INF
    index = 0 # 가장 최단 거리가 짧은 노드(인덱스)
    for i in range(1, n + 1):
        if distance[i] < min_value and not visited[i]:
            min_value = distance[i]
            index = i
    return index

def dijkstra(start):
    # 시작 노드에 대해서 초기화
    # pdb.set_trace()
    distance[start] = 0
    visited[start] = True
    for j in graph[start]:
        distance[j[0]] = j[1]
    # 시작 노드를 제외한 전체 n - 1개의 노드에 대해 반복
    for i in range(n - 1):
        # 현재 최단 거리가 가장 짧은 노드를 꺼내서, 방문 처리
        now = get_smallest_node()
        visited[now] = True
        # 현재 노드와 연결된 다른 노드를 확인
        for j in graph[now]:
            cost = distance[now] + j[1]
            # 현재 노드를 거쳐서 다른 노드로 이동하는 거리가 더 짧은 경우
            if cost < distance[j[0]]:
                distance[j[0]] = cost
````

````python
import heapq
# import sys
# input = sys.stdin.readline
INF = int(1e9) # 무한을 의미하는 값으로 10억을 설정
````

````python
# 모든 간선 정보를 입력받기
for _ in range(m):
    a, b, c = map(int, input().split())
    # a번 노드에서 b번 노드로 가는 비용이 c라는 의미
    graph[a].append((b, c))
````

````python
def dijkstra(start):
    q = []
    # 시작 노드로 가기 위한 최단 경로는 0으로 설정하여, 큐에 삽입
    heapq.heappush(q, (0, start))
    distance[start] = 0
    while q: # 큐가 비어있지 않다면
        # 가장 최단 거리가 짧은 노드에 대한 정보 꺼내기
        print()
        print(q)
        dist, now = heapq.heappop(q)
        print(dist, now)
        print(distance)
        # 현재 노드가 이미 처리된 적이 있는 노드라면 무시
        if distance[now] < dist:
            continue
        # 현재 노드와 연결된 다른 인접한 노드들을 확인
        for i in graph[now]:
            cost = dist + i[1]
            # 현재 노드를 거쳐서, 다른 노드로 이동하는 거리가 더 짧은 경우
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))
                print("push: ", q, cost, i[0])
````

## Floyd-Warshall Algorithm

````python
INF = int(1e9) # 무한을 의미하는 값으로 10억을 설정

# 노드의 개수 및 간선의 개수를 입력받기
n = int(input())
m = int(input())
````

````python
# 2차원 리스트(그래프 표현)를 만들고, 모든 값을 무한으로 초기화
graph = [[INF] * (n + 1) for _ in range(n + 1)]
````

````python
# 자기 자신에서 자기 자신으로 가는 비용은 0으로 초기화
for a in range(1, n + 1):
    for b in range(1, n + 1):
        if a == b:
            graph[a][b] = 0
````

````python
graph
````

````python
# 각 간선에 대한 정보를 입력받아, 그 값으로 초기화
for _ in range(m):
    # A에서 B로 가는 비용은 C라고 설정
    a, b, c = map(int, input().split())
    # 가장 짧은 간선 정보만 저장
    if c < graph[a][b]:
        graph[a][b] = c
````

````python
# 점화식에 따라 플로이드 워셜 알고리즘을 수행
for k in range(1, n + 1):
    for a in range(1, n + 1):
        for b in range(1, n + 1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])
````

````python
graph
````

````python
# 수행된 결과를 출력
for a in range(1, n + 1):
    for b in range(1, n + 1):
        # 도달할 수 없는 경우, 0을 출력
        if graph[a][b] == INF:
            print(0, end=" ")
        # 도달할 수 있는 경우 거리를 출력
        else:
            print(graph[a][b], end=" ")
    print()
````

````python
import heapq

# Colab에서는 이거 쓰면 에러 남
# 백준 제출시 없으면 시간 초과 발생
# import sys
# input = sys.stdin.readline

# 1. 입력
V, E = map(int, input().split()) # 정점, 간선
K = int(input()) # 시작 정점 번호

graph = [[] for _ in range(V+1)]
for _ in range(E):
    u, v, w = map(int, input().split())
    graph[u].append((v, w))
````

````python
graph
````

````python
# 몇 가지 필요한 변수들
INF = 1e9
distance = [INF] * (V+1)
````

````python
# 2. 동작
def dijkstra_self(start):
    priority_q = []
    heapq.heappush(priority_q, (0, start)) # 거리, 정점 번소 순으로 입력
    distance[start] = 0
    while priority_q:
        dist, node = heapq.heappop(priority_q)
        if distance[node] < dist: # 현재 정점을 처리했는지 판별
            continue
        # 그게 아니라면
        for i in graph[node]: # 연결된 점 하나씩 따져보기 i: (정점, 거리)
            # cost = 0
            cost = dist + i[1] # node에 연결된 다음 정점까지의 거리 계산
            if cost < distance[i[0]]: # 다음 정점까지 계산된 거리가 distance에 저장된 최소 거리보다 작으면
                distance[i[0]] = cost # distance의 거리 업데이트
                heapq.heappush(priority_q, (cost, i[0]))
````

````python
dijkstra_self(K)
````
