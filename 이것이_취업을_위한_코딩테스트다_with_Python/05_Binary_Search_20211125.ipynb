## 순차 탐색
- `input().split()`에 공백을 기준으로 여러 값을 넣으면 알아서 리스트로 생성됨!!

```python
def sequential_search(n, target, array):
    for i in range(n):
        if array[i] == target:
            return i + 1

print("원소 개수 입력 후 한 칸 띄고 찾을 문자열 입력")
input_data = input().split() # type(input_data): list
n = int(input_data[0])
target = input_data[1]

print('앞서 입력한 숫자 개수만큼 문자열을 입력, 구분은 띄어쓰기 한 칸')
array = input().split() # type(array): list

# 순차 탐색 수행 결과 출력
print(sequential_search(n, target, array))
```

## 이진 탐색

```python
# 재귀
def binary_search_rec(array, target, start, end):
    # 시작과 끝이 교차된 경우
    if start > end:
        return None
    
    # 중간과 타겟이 일치된 경우
    mid = (start + end) // 2
    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary_search_rec(array, target, start, mid-1)
    else:
        return binary_search_rec(array, target, mid+1, end)
```

```python
# 반복문
def binary_search_rep(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
```

```python
# n(원소 개수)과 target(찾고자 하는 문자열)을 입력받기
n, target = list(map(int, input().split()))
# 전체 원소 입력 받기
array = list(map(int, input().split()))

# 이진 탐색 수행 결과
result = binary_search_rep(array, target, 0, n-1)
if result == None:
    print('X')
else:
    print(result+1)
```
