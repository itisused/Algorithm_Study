## DFS/BFS를 재귀함수로 구현

```python
def dfs(graph, start, visited): # graph: 그래프, v: 시작 노드, visited: 방문 처리
    # v를 Truen 처리
    visited[start] = True
    # # v 노드에 연결되어 있는 것들 찾기
    # temp = graph[v]
    print(v, end=' → ')
    # for temp_ele in temp:
    for i in graph[v]:
        # if not visited[temp_ele]:
        if not visited[i]:
            # dfs(graph, temp_ele, visited)
            dfs(graph, i, visited)
```

```python
from collections import deque

def bfs(graph, start, visited):
    # Queue 구현을 위해 deque 라이브러리 사용
    queue = deque([start])

    # 현재 노드 방문 처리
    visited[start] = True 
    # Queue가 빌 때까지 반복
    while queue:
        # 큐에서 하나의 원소를 뽑아 출력
        v = queue.popleft()
        print(v, end=' → ')
        # 해당 원소왕 연결된, 아직 방문하지 않는 원소들을 Queue에 삽입
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
```

```python
# 각 노드가 연결된 정보를 리스트 자료형으로 표현 (2차원 리스트)
graph = [
    #노드 0,    1,      2,      3,      4,     5,    6,     7,      8
         [], [2,3,8], [1,7], [1,4,5], [3,5], [3,4], [7], [2,6,8], [1,7]
]

# 각 노드가 방문된 정보를 리스트 자료형으로 표현 (1차원 리스트)
visited = [False] * 9 # 혹은 len(graph)로 9 대체

# dfs와 bfs를 한 번에 동시 실행할 경우, dfs를 거쳐 나온 visited가 True로 변경되고
# 초기화를 시키지 않았기 때문에 bfs에 들어가는 visited는 0을 제외한 나머지가
# 모두 True인 것이 들어가서 값이 제대로 안 나온다.

# print(dfs(graph, 1, visited))
print(bfs(graph, 1, visited))
```

```python
# 연결된 것을 찾아야 하기 때문에 DFS

# 1. 입력부
n, m = map(int, input().split(' '))

''' # 예외 처리는 나중에 생각
i = 0
while i < n:
    temp = list(map(int, input()))
    if len(temp) == m:
        case.append(temp)
        i = i + 1
    else:
        i = i - 1
        print('다시!')
        continue
'''
case = []
for i in range(n):
    case.append(list(map(int, input())))
# case

# 2. 동작부
''' 다 아니야!
# DFS 구현
def dfs(graph, start, visited):
    visited[v] = True
    for i in graph[v]:
        if not visited[v]:
            return dfs[graph, i, visited]

# 근데, 주어진 입력값을 그래프로 어떻게 바꾸지?
visited = [[False] * m for _ in range(n)]
case.insert(0, [[0] * m])
start = case[1][1]
'''

# DFS로 특정한 노드를 방문한 뒤에 연결된 모든 노드들 방문(상,하,좌,우)
def dfs(x, y):
    # print(case)
    # pdb.set_trace()
    # 영역 밖을 벗어날 경우 종료
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    
    # 현재 노드를 아직 방문하지 않았다면
    if case[x][y] == 0:
        # 해당 노드 방문 처리
        case[x][y] = 1
        # 상하좌우 위치 재귀로 호출
        dfs(x-1, y)
        dfs(x, y-1)
        dfs(x+1, y)
        dfs(x, y+1)
        return True
    return False

# 모든 노드(위치)에 대해 음료수 채우기
result = 0
for i in range(n):
    for j in range(m):
        # 현재 위치에서 DFS 수행
        if dfs(i, j) == True:
            result = result + 1

print(result)
```

```python
# BFS?
# 최단거리, 없으면 -1
# 1. 입력부
n, m, k, x = map(int, input().split(' '))

# 이게 아니라
# graph = []
# for i in range(m):
#     graph.append(list(map(int, input().split(' '))))

# 이렇게 입력을 받을 것!!! 노드가 연결된 정보를 리스트 자료형으로 표현!!
graph = [[] for _ in range(n+1)] # 인덱스에 맞춰서 0 비움
for _ in range(m):
    a, b = map(int, input().split(' '))
    graph[a].append(b)

# 2. 동작부
from collections import deque

# 2-1) 모든 도시에 대한 최단 거리 초기화
distance = [-1] * (n+1) # 인덱스에 맞춰서 0 비움
distance[x] = 0 # 출발 도시 → 출발 도시의 최단 거리는 0

# 2-2) BFS
queue = deque([x]) #; print(queue); print(list(queue))
while queue:
    now = queue.popleft() # ; print(queue, now); print(distance)
    # 현재 도시에서 이동할 수 있는 모든 도시 확인
    for next_node in graph[now]:
        # 아직 방문하지 않은 도시라면
        if distance[next_node] == -1:
            # 최단 거리 갱신
            distance[next_node] = distance[now] + 1
            queue.append(next_node)
# print(distance)
    
# 최단 거리가 K인 모든 도시의 번호를 오름차순으로 출력
check = False
for i in range(1, n+1):
    if distance[i] == k:
        print(i)
        check = True
    
# 만약 최단 거리가 K인 도시가 없다면, -1 출력
if check == False:
    print(-1)
```

## 기타: DFS와 BFS
- 백준 [1260](https://www.acmicpc.net/problem/1260)

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|----|
