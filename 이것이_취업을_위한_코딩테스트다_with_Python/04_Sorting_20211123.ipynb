## Problem 23: 국영수
- 백준 [10825](https://www.acmicpc.net/problem/10825)

|날짜|소요시간|해설|답안|비고|
|----|----|----|----|----|
|21.11.23|50분 소요|x|x|- __lambda 함수 쓰는 방식 숙지!__ <br> - 백준에서 채점하는데 시간 좀 걸렸음 (속도가 느린가?) <br> - sort, sorted 안쓰고? 정렬이 어디서 쓰인거야? <br> __`import sys`, `input = sys.stdin.readline` 사용할 것!!__|

```python
# 이거 해야 빨라짐!!!
import sys
input = sys.stdin.readline

# 1. 입력부
N = int(input())

stu = []
for _ in range(N):
    name, kor, eng, math = input().split(' ')
    stu.append([name, int(kor), int(eng), int(math)])

# 2. 동작부
stu.sort(key = lambda x: (-x[1], x[2], -x[3], x[0]))

# 3. 출력부
for each_stu in stu:
    print(each_stu[0])
```

```python
# 평균값과의 차이가 가장 작은 것?
# 1. 입력부
N = int(input())
house = list(map(int, input().split()))

house.sort()

avg = sum(house) / len(house)

diff = [abs(each - avg) for each in house]

min_index = diff.index(min(diff))
print(house[min_index])
```

## Problem 25: 실패율
- 프로그래머스 [2019 카카오 신입 공채 1차](https://programmers.co.kr/learn/courses/30/lessons/42889)

|날짜|소요시간|해설|답안|비고|
|----|----|----|----|----|
|21.11.24|20분 초과|x|x|- 처음에는 런타임 오류 발생 → 실패율 계산시 분모가 0이 되는 경우 때문으로 추정됨 <br> - 실패율 계산시 분자가 0인 경우를 별도 처리함으로써 문제 해결|

```python
def solution(N, stages):
    stages.sort()
    through_stages = [0] * (N + 2) # 0부터 N+1까지
    
    # 각 단계별 개수 세기
    for stage in stages:
        through_stages[stage] = through_stages[stage] + 1
    
    fail_rate = [] # 실패율 계산 결과 저장
    answer = [] # 순위
    for idx in range(1, N+1):
        if through_stages[idx] == 0:
            fail_rate.append(through_stages[idx])
        else:
            temp = through_stages[idx] / sum(through_stages[idx:])
            fail_rate.append(temp)

    # 최솟값 찾고, 그 최솟값은 -1로 변경
    # max의 경우 한 개 밖에 못 찾고, 작은 인덱스부터 찾기 때문에 별도의 조치 불필요
    for _ in range(len(fail_rate)):
        idx_result = fail_rate.index(max(fail_rate))
        answer.append(idx_result + 1)
        fail_rate[idx_result] = -1

    return answer
```
