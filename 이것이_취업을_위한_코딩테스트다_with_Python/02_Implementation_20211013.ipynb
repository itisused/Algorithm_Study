## 예제 01: 상하좌우
- 책 110

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|----|
|21.10.13|15분 초과|X||- 책과 풀이 다름|

```python
# 1. 입력부
n_dim = int(input())
direc = list(input().split(' '))

# 2. 동작부
init = [1, 1]
# direc_num = [[0] * 4] # np.zeros 안된다는 가정하에
# L, R, U, D 개수 Count
for ele in direc:
    if ele == 'L':
        if init[1] - 1 > 0:
            init[1] = init[1] - 1
    elif ele == 'R':
        if init[1] + 1 < n_dim:
            init[1] = init[1] + 1
    elif ele == 'U':
        if init[0] - 1 > 0:
            init[0] = init[0] - 1
    elif ele == 'D':
        if init[0] + 1 < n_dim:
            init[0] = init[0] + 1
    else:
        print('Wrong!')
        break

# 3. 출력부
print(init[0]*10 + init[1])
```

|날짜|소요시간|해설|답안|비고|
|:----:|:----:|:----:|:----:|----|

```python
# 1. 입력부
n, m = map(int, input().split(' ')) # 맵의 크기
x, y, d = map(int, input().split(' ')) # 현재 위치와 보고 있는 방향

gone = [[0]*m for _ in range(n)] # 방문 위치
gone[x][y] = 1 # 현재 위치 방문 처리

game_map = []
for idx in range(n):
    # 알아서 m개를 입력할 것이라 가정
    game_map.append(list(map(int, input().split(' '))))
'''
map은 다음과 같이 좌표가 찍힘
(0,0), (0,1), (0,2), ... (0,M)
(1,0), (1,1), (1,2), ...
...
(N,0)                    (N,M)
'''

# 2. 동작부
'''
동서남북으로 정의할 것이 아니라 북서남북으로 정의해야 함
그래야 시계 반대방향으로 회전하는 것을 따지기 유리함
다시... 문제에서 북동남서가 0,1,2,3에 대응되니 이것에 맞춰볼 것
# 동서남북 정의
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]

# 북서남동 정의
dx = [0, -1, 0, 1]
dy = [1, 0, -1, 0]
'''
#북동남서
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 시계 반대 방향 회전 함수 (책)
def CCW_left():
    global d
    d = d - 1
    if d == -1:
        d = 3 # 왜 3이지?

# global을 안 쓸 경우 (나)
def CCW_left2(dd):
    dd = dd - 1
    if d == -1:
        return 3 # 왜 3이지?
    else:
        return dd


# main 부분
cnt = 1 # 처음 시작한 곳을 포함해야 하므로 0이 아닌 1
# init_direct_x = dx[d]
# init_direct_y = dy[d]
turn_time = 0 # 4방향 회전한 것을 체크하기 위한 변수

while True:
    # Step 1) 시계 반대 방향으로 90도 회전
    # d = d - 1 # 잘 못 생각함. d[-1] == d[-5]가 될 줄 알았는데 순환이 아님
    # pdb.set_trace()
    CCW_left()
    # d = CCW_left2(d)
    temp_x = x + dx[d]
    temp_y = y + dy[d]
    
     # 가려는 방향이 땅이라면 & 아직 안 가봤다면 이동할 것
    if game_map[temp_x][temp_y] == 0 and gone[temp_x][temp_y] == 0:
        x = temp_x # x축 이동
        y = temp_y # y축 이동
        gone[x][y] = 1 # 갔다고 표시

        # 여기서부터는 책 따라 함
        cnt = cnt + 1 # 방문한 칸의 개수 증가
        turn_time = 0
        continue
    
    # 가본 곳이거나 바다라면
    else:
        turn_time = turn_time + 1
    
    # 네 방향을 모두 갈 수 없는 경우
    if turn_time == 4:
        temp_x = x - dx[d]
        temp_y = y - dy[d]

        # 뒤로 갈 수 있으면 뒤로 이동
        if game_map[temp_x][temp_y] == 0:
            x = temp_x
            y = temp_y
        # 뒤가 바다라 못 간다면 -> 종료
        else:
            break
        
        # 뒤로 이동했으니 리셋
        turn_time = 0

# 3. 출력부
print(cnt)
```

```python
# 21.09. 언젠가 처음 푼 코드
def solution(s):
    num = len(s)
    max_iter = num // 2

    result = [[], []]
    sep_list = []
    for d in range(1, max_iter+2):
        temp_result = []
        sep_list = [s[d_inn:d_inn+d] for d_inn in range(0, num, d)] # 문자 나누기
        temp_temp = [sep_list.pop(0)]
        
        # 같은 개수 찾기
        for idx, _ in enumerate(sep_list):
            if temp_temp[-1] == sep_list[idx]:    # 같은 문자가 있는 경우
                temp_temp.append(sep_list[idx])
            else:                               # 같은 문자가 없는 경우
                if len(temp_temp) > 1:
                    temp_result.extend([str(len(temp_temp)), temp_temp[-1]])
                else:
                    temp_result.extend([temp_temp[-1]])
                temp_temp.clear()
                temp_temp = [sep_list[idx]]
        
        # 마지막에 작업한게 안 딸려오네
        if len(temp_temp) > 1:
            temp_result.extend([str(len(temp_temp)), temp_temp[-1]])
        else:
            temp_result.extend([temp_temp[-1]])
        
        temp_result = ''.join(temp_result)
        result[0].append(temp_result); result[1].append(len(temp_result))
    
    answer = min(result[1])
    return answer
```

```python
# 여긴 약간의 수학이 있네(?)
# key 리스트(결국, 2차원 행렬)의 90도 회전 함수
def rotation_key(key):
    n_dim = len(key) # key dimension
    rot_key = [[0] * n_dim for _ in range(n_dim)] # 2차원 리스트 생성
    # 여긴 정사각 행렬이니 n_dim 하나로 해결한 것이고
    # 만약 직사각 행렬일 경우 row, column 따로 계산해서 rot_key 초기화해야 함

    for i in range(n_dim):
        for j in range(n_dim):
            rot_key[j][(n_dim-1)-i] = key[i][j]
    return rot_key

# 확대한 lock 리스트(행렬)의 중앙 부분의 각 구성 원소가 모두 1인지 확인
def check_element_one(extend_lock):
    extend_n_dim = len(extend_lock) // 3
    for i in range(extend_n_dim, 2 * extend_n_dim):
        for j in range(extend_n_dim, 2 * extend_n_dim):
            '''
            if extend_lock[i][j] == 1:
                continue
            else:
                return False
            '''
            if extend_lock[i][j] != 1:
                return False
    return True
    

def solution(key, lock):
    
    # 1. lock를 가로 세로 3배씩 확장 (신박하네.)
    lock_dim = len(lock)
    key_dim = len(key)
    extend_lock_dim = lock_dim * 3
    extend_lock = [[0] * extend_lock_dim for _ in range(extend_lock_dim)] # 크기 3배 확장시킨 2차원 리스트 생성
    
    # 2. 확대한 lock의 정중앙에 Origianl lock 삽입
    for i in range(lock_dim, 2*lock_dim):
        for j in range(lock_dim, 2*lock_dim):
            extend_lock[i][j] = lock[i-lock_dim][j-lock_dim]
    
    # 3. key를 회전시켜가면서 풀리는지 확인
    rotation_num = int(360 / 90) # 각 칸이 사각형이라서
    for rotation in range(rotation_num):
        # 먼저 회전
        key = rotation_key(key)
        
        # 훑는다.
        # 어느 M by M 지점에 할지 위치를 잡고
        for i in range(1, extend_lock_dim-key_dim):
            for j in range(1, extend_lock_dim-key_dim):
        # for i in range(extend_lock_dim-key_dim):
        #     for j in range(extend_lock_dim-key_dim):
        # for i in range(lock_dim * 2):
        #     for j in range(lock_dim * 2):

                # 더한다.
                # for k in range(i, i+key_dim):
                #     for l in range(j, j+key_dim):
                        # extend_lock[k][l] = extend_lock[k][l] + lock[k-i][l-j]
                for k in range(key_dim):
                    for l in range(key_dim):
                        extend_lock[i+k][j+l] = extend_lock[i+k][j+l] + key[k][l]
                
                # print(extend_lock)
                # 중앙이 모두 1인지 체크
                answer = check_element_one(extend_lock)
                # pdb.set_trace()
                if answer:
                    return answer
                
                # 다시 원상 복귀. 빼줘야 함
                # for k in range(i, i+key_dim):
                #     for l in range(j, j+key_dim):
                #         extend_lock[k][l] = extend_lock[k][l] - lock[k-i][l-j]
                for k in range(key_dim):
                    for l in range(key_dim):
                        extend_lock[i+k][j+l] = extend_lock[i+k][j+l] - key[k][l]
    return answer

key = [[0, 0, 0], [1, 0, 0], [0, 1, 1]]
lock = [[1, 1, 1], [1, 1, 0], [1, 0, 1]]
print(solution(key, lock))
```
