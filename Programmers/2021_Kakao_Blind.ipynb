- 문자 클래스: [] => [] 사이의 문자들과 매치
	- 범위: - => [a-zA-Z] 모든 알파벳, [0-9] 모든 숫자 [ㄱ-ㅎㅏ-ㅣ] 모든 한글?
	- 반대: ^ => [^0-9] 숫자가 아닌 문자만
- a.b => a + 모든 문자 1개 이상 + b
- a[.]b => a.b 인 경우
- 반복: `*` => `ca*t`: ct부터 caaaaaa...aaat까지 a가 0번부터 무한대까지 반복 가능
- 반복: + => ca+t a가 1번 이상인 경우(*와 나머지는 다 동일)
- 반복: {m,n} => ca{m}t, ca{m,n}t, ca{0,}t, ca{1,}t 등 m회, m부터 n회 반복, *랑 - 동일, +랑 동일
- 있거나 없거나: ? ab?c == abc or ac
- 또는: | => a|b
- 시작: ^ => ^abc : 시작이 abc 인 경우, 중간에 있음 해당 안됨
- 끝: `$` => `abc$` : 끝이 abc인 경우
- 공백: \b => r'\bclass\b': no class at (ok), subclass is (not ok)
- r'~~': Raw string (위가 예시)
- re.compile('정규식'); re.match('문자') == re.match('정규식', '문자')
- grouping....: ()로 묶기

```python
import re

def solution(new_id):
    answer = ''
    
    answer = new_id.lower()
    answer = re.sub('[^a-z0-9-_.]', '', answer)
    answer = re.sub('[.]+', '.', answer)
    answer = re.sub('^[.]|[.]$', '', answer)
    if len(answer) == 0:
        answer = 'a'
    if len(answer) >= 16:
        answer = answer[:15]
    answer = re.sub('[.]$', '', answer)
    if len(answer) <= 2:
        answer = answer + answer[-1] * (3 - len(answer))
    
    return answer
```
