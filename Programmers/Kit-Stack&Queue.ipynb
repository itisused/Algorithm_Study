||2) 실패 방식|잔여 일을 계산한 리스트의 인접한 원소 사이의 대소비교를 통해 '(이전 원소)<(다음 원소)'를 기준으로 리스트를 나누고 구간별 숫자의 개수 Count|
||3) 풀이 방식|별도의 리스트에 값을 저장하되 해당 리스트 내 첫 번째 원소와 잔여 일을 계산한 리스트의 첫 번째 원소와의 대소 비교를 통해 '(별도 리스트 첫 번째 원소)>&=(잔여 일수 리스트의 첫 번째 원소)'가 나오는 경우만 저장. '<'의 경우가 나오는 경우 별도 리스트의 개수를 세서 저장하고, 해당 리스트를 비움|
||4) 추가 입출력|(Q&A 참고) progresses: [96, 99, 98, 97] speeds: [1,1,1,1] answer: [4]||

```python
# Try 1 (2021-09-13 실패)
def solution(progresses, speeds):
    num = len(progresses)
    day = []

    end_val = 100
    for pro, spe in zip(progresses, speeds):
        temp = 0
        temp = (end_val - pro) // spe
        if (end_val - pro) % spe == 0:
            day.append(temp)
        else:
            day.append(temp+1)
    
    temp_list = []
    answer = []
    # pdb.set_trace()
    temp_list.append(day.pop(0))
    for ele in day:
        if ele < temp_list[0]:
            temp_list.append(ele)
        else:
            answer.append(len(temp_list))
            temp_list.clear()
            temp_list.append(ele)
    answer.append(len(temp_list))
    
    return answer
```

```python
def solution(progresses, speeds):
    num = len(progresses)
    day = []

    end_val = 100
    for pro, spe in zip(progresses, speeds):
        temp = 0
        temp = (end_val - pro) // spe # 몫
        if (end_val - pro) % spe == 0:
            day.append(temp)
        else:
            day.append(temp+1)
    
    temp_stack = [day.pop(0)]
    answer = []
    while day:
        if temp_stack[0] >= day[0]:
            temp_stack.append(day.pop(0))
        else:
            answer.append(len(temp_stack))
            temp_stack = [day.pop(0)]
    answer.append(len(temp_stack))
            
    return answer
```
