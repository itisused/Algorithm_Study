```python
def solution(numbers):
    ref = [1,2,3,4,5,6,7,8,9]
    '''
    # 런타임 에러 발생(한 경우만) 왜???
    for n in numbers:
        ref[ref.index(int(n))] = 0
    answer = sum(ref)
    '''
    sum_ref = sum(ref)
    for n in numbers:
        sum_ref = sum_ref - int(n)
    answer = sum_ref
    return answer
```
