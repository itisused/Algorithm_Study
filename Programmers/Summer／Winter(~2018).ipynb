```python
from itertools import combinations
import math

def prime_num(n):
    # 기본 버전
    # for i in range(2, n):
    # 개선된 버전
    for i in range(2, int(math.sqrt(n))+1):
        if n % i == 0:
            return False
    return True

def solution(nums):
    # 경우의 수 조합
    list_comb = list(combinations(nums, 3))
    cnt = 0
    for comb in list_comb:
        temp = sum(comb)
        # 소수 판별 필요
        if prime_num(temp):
            cnt = cnt + 1
    
    answer = cnt
    return answer
```
