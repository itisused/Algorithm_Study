||3) 풀이 방식|Array로 변환하여 해결(세부 동작 기억 안남)|
||4) 그 외|시험 당일 해결했으나, 시간을 너무 많이 잡아먹음|

```python
# Try 1: 시간 초과 4건 /24건

import pandas as pd
import numpy as np

def solution(id_list, report, k):

    # 중복 신고 제거
    report = list(set(report))

    # Report 분할
    new_report = [str(case).split(sep=' ') for case in report]

    # Dataframe 생성
    num = len(id_list)
    init_val = np.zeros((num, num), dtype=int)

    df = pd.DataFrame(init_val, index=id_list, columns=id_list)

    for each in new_report:
        df.loc[each[0], each[1]] = df.loc[each[0], each[1]] + 1
    
    # sum 행, mail 열 생성
    df.loc['sum'] = 0
    df['mail'] = 0
    
    # 신고 횟수 합산
    for idx, id in enumerate(id_list):
        df.iloc[-1, idx] = sum(df[id])
    
    # k 이상 경고 받은 자 추출
    temp = []
    for idx, id in enumerate(id_list):
        if df.iloc[-1, idx] >= k:
            temp.append(id)
    
    # 중지시킬 사람을 신고한 사람
    # send_id = []
    for each in new_report:
        if each[1] in temp:
            df.loc[each[0], 'mail'] = df.loc[each[0], 'mail'] + 1

    answer = list(df['mail'])[:-1]
    return answer
```

```python
# Try 3: Submit

import numpy as np

def solution2(id_list, report, k):

    # 중복 신고 제거
    report = list(set(report))

    # Report 분할
    new_report = [str(case).split(sep=' ') for case in report]

    # 숫자로 변환
    id_list_num = [id_list.index(name) for name in id_list]
    
    id_list_dict = {name:id_list.index(name) for name in id_list}
    new_report_num = [(id_list_dict[case[0]], id_list_dict[case[1]]) for case in new_report]

    # Array 생성
    num = len(id_list)
    data_sample = np.zeros((num+1, num+1), dtype=int)
    # dtype=int 넣으면 'Object of type int64 is not JSON serializable' 오류남
    # dtype=int를 제거하니 제대로 동작함

    # Array에 표기
    for each_num in new_report_num:
        data_sample[each_num[0], each_num[1]] = data_sample[each_num[0], each_num[1]] + 1
    
    # 신고 횟수 합산
    for idx, id in enumerate(id_list_num):
        data_sample[-1, idx] = sum(data_sample[:-1, id])
    
    # k 이상 경고 받은 자 추출
    temp = []
    for idx, id in enumerate(id_list_num):
        if data_sample[-1, idx] >= k:
            temp.append(id)
    
    # 중지시킬 사람을 신고한 사람
    for each in new_report_num:
        if each[1] in temp:
            data_sample[each[0], -1] = data_sample[each[0], -1] + 1

    answer = list(data_sample[:-1, -1])
    return answer
```

### 2) [k 진수에서 소수의 개수 구하기](https://programmers.co.kr/learn/courses/30/lessons/92335) (Lv. 2)

||3) 그 외|시험 당일 해결 <br> 검색의 도움을 받음|

```python
import numpy as np

# 소수 판별
def p_num(num):
    if num < 2:
        return 0
    max_iter = int(np.sqrt(num))
    for i in range(2, max_iter+1):
        if num % i == 0:
            return 0
    return 1

def solution(n, k):

    # 10진수 -> k진수
    # pdb.set_trace()
    if k == 10:
        n_base_k = n
    else:
        n_base_k = ''
        while n > 0:
            n, spa = divmod(n, k)
            n_base_k = n_base_k + str(spa)
        n_base_k = int(n_base_k[::-1])

    # pdb.set_trace()
    # 0을 기준으로 분할
    str_n_base_k = str(n_base_k)
    str_n_base_k_split = str_n_base_k.split('0')

    # 공백 제거
    num_empty = str_n_base_k_split.count('')
    for i in range(num_empty):
        str_n_base_k_split.remove('')

    # 소수 개수 계산
    answer = 0
    for n in str_n_base_k_split:
        answer = answer + p_num(int(n))

    return answer

# print(solution(110011, 10))
```
