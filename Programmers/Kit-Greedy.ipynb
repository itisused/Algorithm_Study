### 1) [체육복](https://programmers.co.kr/learn/courses/30/lessons/42862) (Lv. 1)

|날짜|비고||
|--|--|--|
|2021.11.13|1) 개념|완전 탐색?|
||2) 풀이 방식|모든 경우 확인 후 결정|

```python
def solution(n, lost, reserve):
    # lost 앞 뒤에 reserve가 있는지 확인해야 함
    cloth = [1] * (n+1) # 입력 값과 리스트 인덱스를 일치시키기 위해 인덱스 1부터 처리하려고 n+1
    cloth[0] = 0
    
    # 1. 옷 개수 확인
    answer = 0 # 학생 수 Count
    for i in range(1, n+1):
        if i in lost and i not in reserve:
            cloth[i] = cloth[i] - 1
        elif i not in lost and i in reserve:
            cloth[i] = cloth[i] + 1
        else:
            cloth[i] = cloth[i]
    
    # 2. 한 명씩 탐색
    for i in range(1, n+1):
        if cloth[i] > 0:
            answer = answer + 1
        elif cloth[i] == 0:
            if i == 1 and cloth[i+1] == 2:
                cloth[i] = cloth[i] + 1
                cloth[i+1] = cloth[i+1] - 1
                answer = answer + 1
            elif (i > 1 and i < n) and cloth[i-1] == 2:
                cloth[i] = cloth[i] + 1
                cloth[i-1] = cloth[i-1] - 1
                answer = answer + 1
            elif (i > 1 and i < n) and cloth[i+1] == 2:
                cloth[i] = cloth[i] + 1
                cloth[i+1] = cloth[i+1] - 1
                answer = answer + 1
            elif i == n and cloth[i-1] == 2:
                cloth[i] = cloth[i] + 1
                cloth[i-1] = cloth[i-1] - 1
                answer = answer + 1
    return answer
```

```python
def check_reverse(ascii_list):
    # 좌우로 넘기는게 문제네 뭐가 더 간단한지...
    # 1. ??...?/00...0/??...? 이런 구조로 0이 덩어리로 뭉쳐서 3등분으로 나눌 수 있으면
    # 2. 앞쪽 ?의 개수가 0 덩어리 내 0 개수보다 적은 경우 왼쪽으로 넘어가서 마지막으로 이동하는 것이 더 효율적
    # (추가) 
    # 정방향 진행 True
    # 중간에 방향 바꿔야하면 False
    # temp = 0
    temp_list = []
    for idx in range(len(ascii_list) - 1):
        if (ascii_list[idx] > 0 and ascii_list[idx + 1] == 0) or (ascii_list[idx] == 0 and ascii_list[idx + 1] > 0):
            # temp = temp + 1
            temp_list.append(idx)
    print(temp_list)
    
    if len(temp_list) == 2 and temp_list[0] + 1 <= temp_list[1] - temp_list[0]:
        return False, temp_list
    else:
        return True, temp_list

def solution(name):
    # ASCII 이용? ㅇㅇ
    # 상하로 움직이는 것도 ASCII로 할 경우 어떻게 따지지?
    # 1. 전체 26자 중 A가 기준이니까 얘를 기준으로 생각하면,
    # 2. 만약 A의 인덱스가 0이라면 인덱스 13에 해당하는 N을 기준으로 N 보다 앞에 글자는 ↑로 이동
    # 3. N 보다 뒤에 있는 글자는 ↓로 이동, N은 어디로 이동하든 상관 없음 (N은 13번 움직여야 함)
    # ASCII를 쓰고 A의 ASCII 값을 빼주면, 해당 인덱스 값이 A부터 ↑로 이동하는 횟수에 해당함
    
    name_ascii = []
    start_ascii = ord('A')
    end_change_ascii = ord('Z') - start_ascii
    middle_change_ascii = ord('N') - start_ascii
    for ch in name:
        # name_ascii.append(ord(ch) - start_ascii)
        change_ascii = ord(ch) - start_ascii
        if change_ascii > middle_change_ascii:
            # ↓로 이동하는 횟수로 변환 Z인 25의 경우 1로 변경, O는 12로 변경
            change_ascii = end_change_ascii - change_ascii + 1
        name_ascii.append(change_ascii)
    
    # 어떻게 이동할지 결정
    TF, idx_list = check_reverse(name_ascii)
    # print(TF, idx_list, sum(name_ascii), len(name_ascii))
    if TF:
        # 정방향
        # print('here')
        return sum(name_ascii) + len(name_ascii) - 1
    else:
        # 역방향
        # print('Here')
        return sum(name_ascii) + idx_list[0] * 2 + len(name_ascii) - idx_list[1] - 1
```
