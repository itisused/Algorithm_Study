### 1) [완주하지 못한 선수](https://programmers.co.kr/learn/courses/30/lessons/42576) (Lv. 1)

|날짜|비고||
|--|--|--|
|2021.11.13|1) 개념|해시가 어디에?|
||2) 풀이 방식|정렬 후 순서대로 단순 비교 <br> 효율성 점수도 있음|

```python
def solution(participant, completion):
    participant.sort()
    completion.sort()
    
    for i in range(len(completion)):
        if participant[i] != completion[i]:
            return participant[i]
    return participant[i+1]
```
