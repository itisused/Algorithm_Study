- 해결 방안... ([인터넷 검색](https://stackoverflow.com/questions/6473679/transpose-list-of-lists]))
    - `board_T = list(map(list, zip(*board)))`
    - **반드시 board 앞에 *을 붙여줘야 함!!**
    - 이해는 명확하지 않음, `for a, b in zip(A, B)`에 쓰이는 zip과 동일하다는데...
    - 그 외 참고: [참고1](https://chancoding.tistory.com/144), [참고2](https://codingpractices.tistory.com/50?category=1026198)

````python
def solution(board, moves):
    crane = 0
    # 1. 세로로 배열 변경
    board_T = list(map(list, zip(*board)))
    cnt = 0 # 터진 횟수
    result = []
    for n in moves:
        # 해당 칸이 비어 있는 경우 넘어갈 것
        if len(board_T[n-1]) == 0:
            continue
        
        # 2. 해당 칸에 인형이 있다면
        while True: # 0 제거
            crane = board_T[n-1].pop(0)
            if crane:
                break
        
        # 만약 바구니에 인형이 없다면 그냥 넣고
        if len(result) == 0:
            result.append(crane)
            crane = 0
            continue
        
        # 바구니에 인형 있다면
        if crane == result[-1]:
            cnt = cnt + 1 # 터진 횟수 증가
            result.pop() # 터뜨리기
        else:
            result.append(crane)
        crane = 0
            
    answer = cnt * 2 # 터진 횟수가 아니라 사라진 인형 개수이므로
    return answer
````
