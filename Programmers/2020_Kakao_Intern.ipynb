||3) 풀이 방식|(해설 참고)좌표 설정하고 맨허튼 거리 계산|
||4) 그 외|- `abs()` 내장 함수 <br> - 0번 처리, 일부 자잘한 것에서 오류가 있었음, 생각 좀 잘|

```python
def man_dist(pos, l, r, h):
    l_dist = abs(pos[0] - l[0]) + abs(pos[1] - l[1])
    r_dist = abs(pos[0] - r[0]) + abs(pos[1] - r[1])
    if l_dist > r_dist:
        return 'R'
    elif l_dist < r_dist:
        return 'L'
    else:
        if h == 'left':
            return 'L'
        else:
            return 'R'

def solution(numbers, hand):
    position = [
        [3,1],
        [0,0], [0,1], [0,2],
        [1,0], [1,1], [1,2],
        [2,0], [2,1], [2,2],
        [3,0], [3,2]
    ]
    
    l_hand = position[10]
    r_hand = position[11]
    
    answer = ''
    for n in numbers:
        if n % 3 == 1:
            l_hand = position[n]
            answer = answer + 'L'
        elif n % 3 == 0 and n != 0:
            r_hand = position[n]
            answer = answer + 'R'
        else:
            ran_hand = man_dist(position[n], l_hand, r_hand, hand)
            if ran_hand == 'L':
                l_hand = position[n]
            else:
                r_hand = position[n]
            answer = answer + ran_hand
    
    return answer
```
