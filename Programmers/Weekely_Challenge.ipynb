```python
import numpy as np
import pdb

def solution(scores):
    avg_scores = []
    grade = []

    # 계산을 열끼리 하기 때문에 표 전치(Transpose)와 리스트 변환
    scores = np.transpose(scores).tolist()
    for idx, row_scores in enumerate(scores):
        # pdb.set_trace()
        # 자기자신이 유일한 최대/최소인지 판단
        self_score = row_scores[idx]
        if self_score == np.max(row_scores) or self_score == np.min(row_scores):
            c_row_scores = row_scores
            del c_row_scores[idx]
            if self_score in c_row_scores:
                avg_scores.append(np.mean(row_scores))
            else:
                avg_scores.append(np.mean(c_row_scores))
        else:
            avg_scores.append(np.mean(row_scores))
        # pdb.set_trace()

        # 학점 구하기
        if avg_scores[idx] >= 90 and avg_scores[idx] <= 100:
            grade.append('A')
        elif avg_scores[idx] >= 80 and avg_scores[idx] < 90:
            grade.append('B')
        elif avg_scores[idx] >= 70 and avg_scores[idx] < 80:
            grade.append('C')
        elif avg_scores[idx] >= 50 and avg_scores[idx] < 70:
            grade.append('D')
        else:
            grade.append('F')

    answer = ''.join(grade)
    return answer
```
