### 1) [로또의 최고 순위와 최저 순위](https://programmers.co.kr/learn/courses/30/lessons/77484) (Lv. 1)

|날짜|비고||
|--|--|--|
|2022-03-09|1) 개념|?|
||2) 실패 방식|-|
||3) 풀이 방식|List 내 원소 개수 Count 하는 count 사용|
||4) 그 외|List 관련 내부 함수 숙지 좀|

```python
def solution(lottos, win_nums):
    # 1. 정렬
    lottos = sorted(lottos)
    win_nums.sort()
    
    # 2. lottos 0 개수, win_nums와 일치 개수
    zero = lottos.count(0) # 0의 개수
    
    count = 0 # win_nums와 일치 개수
    for lot in lottos[zero:]:
        if lot in win_nums:
            count += 1 # count = count + 1
    
    # 3. 순위 판정
    same_max = zero + count
    same_min = count
    
    answer = []
    if 7-same_max < 6:
        answer.append(7-same_max)
    else:
        answer.append(6)
    
    if 7-same_min < 6:
        answer.append(7-same_min)
    else:
        answer.append(6)
    
    return answer
```
