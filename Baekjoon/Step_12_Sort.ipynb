### 5) [소트인사이트](https://www.acmicpc.net/problem/1427)
- 백준을 너무 오랜만에 접해봐서 프로그래머스 형식으로 풀어버림
  - Backjoon은 입력부(?) 필수 / Programmers는 함수만 만들면 끝
  - Backjoon은 외부 라이브러리 사용 불가 (대표적: numpy)
- 주어진 문제는 python의 sort 함수를 통해 간단히 완성 가능함

```python
def solution():
    num_list = list(input())
    answer = []
    for i in range(len(num_list)):
        temp = max(num_list)
        ind = num_list.index(temp)
        answer.append(num_list.pop(ind))
    return ''.join(answer)

print(solution())
```
