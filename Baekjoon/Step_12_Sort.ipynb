### 5) [소트인사이트](https://www.acmicpc.net/problem/1427)
- 백준을 너무 오랜만에 접해봐서 프로그래머스 형식으로 풀어버림

```python
def solution():
    num_list = list(input())
    answer = []
    for i in range(len(num_list)):
        temp = max(num_list)
        ind = num_list.index(temp)
        answer.append(num_list.pop(ind))
    return ''.join(answer)

print(solution())
```
