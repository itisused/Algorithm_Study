- Try 1: 런타임 에러

```python
# 1. 입력
num = list(map(int, input().split()))
num

# 2. 동작
result = ''
i = 9
while sum(num) > 0:
    # 만약 해당 숫자가 없는 경우 넘어간다.
    if num[i] == 0:
        i = i - 1
        continue
    # 만약 마지막 숫자와 지금 넣으려는 숫자가 같지 않을 경우
    if result[-1] != str(i):
        # 숫자를 합치고, 리스트에서 값을 1 뺌
        result = result + str(i)
        num[i] = num[i] - 1
    # 만약 같다면 그 다음 0이 아닌 번호를 찾아야 함
    else:
        i = 9 # 초기화?

# 3. 출력
if result[0] == 0:
    print(0)
else:
    print(int(result))
```

```python
# 1. 입력
num = list(map(int, input().split()))

# 2. 동작
result = [0]
i = 9
while sum(num) > 0:
    # pdb.set_trace()
    # 만약 해당 숫자가 없는 경우 넘어간다.
    if num[i] == 0:
        i = i - 1
        continue
    # 만약 마지막 숫자와 지금 넣으려는 숫자가 같지 않을 경우
    # if result[-1] != str(i):
    if result[-1] != i:
        # 숫자를 합치고, 리스트에서 값을 1 뺌
        # result = result + str(i)
        result.append(i)
        num[i] = num[i] - 1
    # 만약 같다면 그 다음 0이 아닌 번호를 찾아야 함
    else:
        i = 9 # 초기화?

result.pop[0]
# 3. 출력
if result[0] == 0:
    print(0)
else:
    print(int(result))
```
