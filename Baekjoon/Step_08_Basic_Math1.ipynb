### 2) [벌집](https://www.acmicpc.net/problem/2292)

```python
N = int(input())

# 범위 찾기
# 끝 항: a_n = 1 + 3n(n-1)
n = 1; a_n = 0
while True:
    a_n = int(1 + n*(n-1)*3)
    if a_n >= N:
        break
    n = n + 1
print(n)
```

### 3) [분수찾기](https://www.acmicpc.net/problem/1193)
- 너무 복잡하게 생각

```python
X = int(input())

# 끝 항 번호 점화식: a_n = n(n+1)/2
n = 1
a_n = 0
while True:
    a_n = int(n * (n+1) / 2)
    if a_n >= X:
        break
    n = n + 1

# X가 n번째 그룹이라면 분자+분모 = n+1
# n이 홀수: 분모는 n부터 --, 분자는 1부터 ++
# n이 짝수: 분모는 1부터 ++, 분자는 n부터 --
# 증감은 a_n - X 만큼
d = a_n - X
if n % 2 == 0:
    mother = 1 + d
    son = n - d
    print(son, '/', mother, sep='')
else:
    mother = n - d
    son = 1 + d
    print(son, '/', mother, sep='')
```

### 5) [ACM 호텔](https://www.acmicpc.net/problem/10250)

```python
T = int(input())

result = []
for test_case in range(1, T+1):
    H, W, N = map(int, input().split())
    floor = N % H
    num = N // H + 1
    if floor == 0:
        floor = H
        num = num - 1

    if num < 10:
        temp = str(floor) + '0' + str(num)
    else:
        temp = str(floor) + str(num)
    result.append(temp)

for output in result:
    print(output)
```
