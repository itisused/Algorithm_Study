### 3) [알파벳 찾기](https://www.acmicpc.net/problem/10809)

- `from string import ascii_lowercase`: 알파벳 string 형태로 Loading

````python
# import sys
# input = sys.stdin.readline

from string import ascii_lowercase

S = list(input())
ref = list(ascii_lowercase)

for alp in ref:
    if alp in S:
        print(S.index(alp), end = ' ')
    else:
        print(-1, end = ' ')
````

### 4) [문자열 반복](https://www.acmicpc.net/problem/2675)

````python
# import sys
# input = sys.stdin.readline

# 1. 입력
T = int(input()) # Test Case
string = []
for _ in range(T):
    R, S = input().split()
    string.append([int(R), S])

# 2. 동작 및 출력
for string_ele in string:
    r = string_ele[0]
    each_S = string_ele[1]
    
    new_S = ''
    for s in each_S:
        new_S = new_S + s * r
    
    print(new_S)
````

### 5) [단어 공부](https://www.acmicpc.net/problem/1157)
- `dictionary.items()`: Dictionary의 key, value를 모두 갖고옴

|날짜|비고|
|:-:|--|
|22.01.21|1) 틀렸음... 어디서? 왜?? <br> 2) 공백이 입력되는 경우? 출력 보완 → 그래도 틀렸음 <br> 3) 아무 입력이 없을 때? → 틀렸음 <br> 4) 공백 다음에 문자 넣을 때? → 틀렸음 <br><br> 모두 채점 진행이 100%까지 갔는데 최종적으로 틀림 <br> 일단 포기|


````python
# import sys
# input = sys.stdin.readline

from string import ascii_lowercase

# 1. 입력
word = input()
word = list(word.lower())

# 2. 동작
char_dic = {}
for w in word:
    if w == ' ' or w == '':
        continue
    
    if w in char_dic:
        char_dic[w] = char_dic[w] + 1
    else:
        char_dic[w] = 1

# 최댓값
if len(char_dic) == 0:
    print('?')
else:
    max_value = max(list(char_dic.values()))
    max_char = []
    for c, v in char_dic.items(): # c ← char_dic.keys(), v ← char_dic.values()
        if v == max_value:
            max_char.append(c)

    # 3. 출력
    if len(max_char) == 1:
        print(max_char[0].upper())
    else:
        print('?')
````

### 7) [상수](https://www.acmicpc.net/problem/2908)

````python
# 1. 입력
num_set = list(input().split())

# 2. 동작
change_set = []
for n_set in num_set:
    temp_n = ''
    for n_char in n_set:
        temp_n = n_char + temp_n # 순서 뒤집기
    change_set.append(int(temp_n))

print(max(change_set))
````

````python
# up_word = list(input().upper())
up_word = input()

# 동작
# 다이얼이 다시 원위치로 돌아가는데 걸리는 시간은 무시
dial = {
    2:['A','B','C'], 3:['D','E','F'], 4:['G','H','I'], 5:['J','K','L'],
    6:['M','N','O'], 7:['P','Q','R','S'], 8:['T','U','V'], 9:['W','X','Y','Z']
}
dial_time = 0
for u_w in up_word:
    for dial_k, dial_v in dial.items():
        if u_w in dial_v:
            dial_time = dial_time + dial_k + 1

print(dial_time)
````
